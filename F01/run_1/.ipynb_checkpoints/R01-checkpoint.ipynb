```python
nuc_list_in = ["238pu", "239pu", "240pu", "241pu", "242pu", "241am"]

def get_ev(file):
    db = cym.dbopen(file)
    ev = cym.Evaluator(db)
    return ev
```

```python
# Get Beginning of Cycle Fractions

def get_nuc_frac_boc(ev, nuc):
    df = tm.transactions(ev, receivers=['lwr'], nucs=[nuc] )
    df = df[df['Mass'] > 0 ]

    tot = tm.transactions(ev, receivers=['lwr'], nucs=nuc_list_in) 
    tot = tot[tot['Mass'] > 0 ]                

    df['Mass'] = df['Mass']/tot['Mass']
    return df.iloc[2]['Mass']

def get_pu_frac_boc(ev):
    df = tm.transactions(ev, receivers=['lwr'], nucs=nuc_list_in )
    df = df[df['Mass'] > 0 ]

    tot = tm.transactions(ev, receivers=['lwr']) 
    tot = tot[tot['Mass'] > 0 ]                

    df['Mass'] = df['Mass']/tot['Mass']
    return df.iloc[2]['Mass']

def boc_pu_compo(ev):
    compo = []
    for nuc in nuc_list_in:
        compo.append(get_nuc_frac_boc(ev, nuc))
    return compo

def get_boc_intel(ev):
    tt = boc_pu_compo(ev) +  [get_pu_frac_boc(ev)]
    return tt
```

```python
# Get End of Cycle Fractions

def get_nuc_frac_boc(ev, nuc):
    df = tm.transactions(ev, receivers=['lwr'], nucs=[nuc] )
    df = df[df['Mass'] > 0 ]

    tot = tm.transactions(ev, receivers=['lwr'], nucs=nuc_list_in) 
    tot = tot[tot['Mass'] > 0 ]                

    df['Mass'] = df['Mass']/tot['Mass']
    return df.iloc[2]['Mass']

def get_pu_frac_boc(ev):
    df = tm.transactions(ev, receivers=['lwr'], nucs=nuc_list_in )
    df = df[df['Mass'] > 0 ]

    tot = tm.transactions(ev, receivers=['lwr']) 
    tot = tot[tot['Mass'] > 0 ]                

    df['Mass'] = df['Mass']/tot['Mass']
    return df.iloc[2]['Mass']

def boc_pu_compo(ev):
    compo = []
    for nuc in nuc_list_in:
        compo.append(get_nuc_frac_boc(ev, nuc))
    return compo

def get_boc_intel(ev):
    tt = boc_pu_compo(ev) +  [get_pu_frac_boc(ev)]
    return tt
```

```python
def get_fraction_boc(files):
    data =[]
    for file in glob.glob(files):
        ev = get_ev(file)
        data.append(get_boc_intel(ev))
    return data
```
